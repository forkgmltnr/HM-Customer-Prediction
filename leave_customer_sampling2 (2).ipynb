```python
df_clothes_no_sample.sort_values(['customer_label', 't_dat'])
df_clothes_no_sample.reset_index(inplace = True, drop = True)
df_clothes_no_sample.info()
```

```python
df_clothes_no_sample.sort_values(['customer_label', 't_dat'], inplace=True)

from tqdm import tqdm

# t_dat을 날짜 형식으로 변환
df_clothes_no_sample['t_dat'] = pd.to_datetime(df_clothes_no_sample['t_dat'])

# purchase_diff 컬럼 추가 및 초기화
df_clothes_no_sample['purchase_diff'] = None

# 이전 구매 날짜를 저장할 딕셔너리 초기화
prev_purchase_date = {}

# purchase_diff 계산을 위한 함수 정의
def calculate_purchase_diff(row):
    customer_label = row['customer_label']
    current_date = row['t_dat']
    
    # 이전 구매 날짜가 있는 경우에만 계산 수행
    if customer_label in prev_purchase_date:
        prev_date = prev_purchase_date[customer_label]
        diff = (current_date - prev_date).days
        
        return diff
    
    # 이전 구매 날짜가 없거나, 유효한 구매 주기를 계산할 수 없는 경우
    return None

# purchase_diff 컬럼 추가 및 함수 적용 (진행 상황 표시)
with tqdm(total=len(df_clothes_no_sample)) as pbar:
    for idx, row in df_clothes_no_sample.iterrows():
        customer_label = row['customer_label']
        current_date = row['t_dat']

        diff_value = calculate_purchase_diff(row)
        
        if pd.notnull(diff_value):
            df_clothes_no_sample.at[idx, 'purchase_diff'] = diff_value
        
        if pd.notnull(row['purchase_diff']):
            prev_purchase_date[customer_label] = current_date
        
        # 첫 번째로 등장하는 고객의 첫 번째 구매 날짜 정보 초기화
        if customer_label not in prev_purchase_date:
            df_clothes_no_sample.at[idx, 'purchase_diff'] = np.nan
            prev_purchase_date[customer_label] = current_date
            
        pbar.update(1)

df_clothes_no_sample.head(5)
```

# 고객 구매 이력 데이터 추가

```python
def compute_purchase_diff_for_clothes(df):
    df.sort_values(['customer_label','t_dat'], inplace=True)# 데이터를 customer_label, t_dat 기준 정렬
    
    df['t_dat'] = pd.to_datetime(df['t_dat'])# t_dat 컬럼 날짜 형식으로 변환
    
    df['purchase_diff'] = df.groupby('customer_label')['t_dat'].diff().dt.days
    #각 고객별로 t_dat 차이를 계산하여  purchase_diff 컬럼에 저장한다.
    #이 값은 구매한 날짜와 이전 구매 날짜 사이의 일수를 나타낸다.
    #diff 함수는 현재 행의 값과 이전 행의값 차이를 계산한다. 여기선 고객별로 t_Dat 차이 계산
    
    
    df['purchase_diff'] = df['purchase_diff'].replace(0, np.nan)
    #purchase_diff 값이 0인 경우 NaN
    
    df['purchase_diff'] = df.groupby(['customer_label','t_dat'])['purchase_diff'].transform('first')
    # 고객이 같은 날 여러번 구매한 경우purchase_diff 값을 동일하게 만든다.
    
    return df

df_clothes_no_sample_updated = compute_purchase_diff_for_clothes(df_clothes_no_sample.copy())
```

```python
# NaN값이 엉뚱하게 출력된다.
specific_customer = df_clothes_no_sample_updated[df_clothes_no_sample_updated['customer_label'] == 4]
specific_customer = specific_customer[['t_dat', 'purchase_diff']]
specific_customer.head(10)
```

```python
def compute_purchase_diff_final_v5(df):
    df.sort_values(['customer_label', 't_dat'], inplace=True)
    df['t_dat'] = pd.to_datetime(df['t_dat'])
    

    df['purchase_diff'] = df.groupby('customer_label')['t_dat'].diff().dt.days
    
  
    mask_same_date = df.duplicated(subset=['customer_label', 't_dat'], keep='first')
    #첫 구매와 동일한 날짜에 이루어진 중복 구매를 찾아내는 불리언 마스크 생성
    # keep='first' 첫번째를 제외한 나머지 모든 중복 항목을 True 표시
   
    df.loc[mask_same_date, 'purchase_diff'] = 0
    # 첫 구매와 동일한 날짜에 이루어진 중복 구매의 'purchase_diff' 값을 0으로 설정
   
    df['purchase_diff'] = df.groupby(['customer_label', 't_dat'])['purchase_diff'].transform('first')
    # 첫 구매와 동일한 날짜에 이루어진 모든 구매의 'purchase_diff' 값을 첫 구매의 'purchase_diff' 값으로 설정
    return df


df_sample_final_v5 = compute_purchase_diff_final_v5(df_clothes_no_sample_updated.copy())
```

```python
def compute_purchase_diff_final_v6(df):
    df.sort_values(['customer_label', 't_dat'], inplace=True)
    df['t_dat'] = pd.to_datetime(df['t_dat'])
    

    df['purchase_diff'] = df.groupby('customer_label')['t_dat'].diff().dt.days
    

    mask_same_date = df.duplicated(subset=['customer_label', 't_dat'], keep='first')
    

    df.loc[mask_same_date, 'purchase_diff'] = 0
    # 첫 구매와 동일한 날짜에 이루어진 중복 구매의 'purchase_diff' 값을 0으로 설정

    
    df['purchase_diff'] = df.groupby(['customer_label', 't_dat'])['purchase_diff'].transform('first')
    #  첫 구매와 동일한 날짜에 이루어진 모든 구매의 'purchase_diff' 값을 첫 구매의 'purchase_diff' 값으로 설정

    mask_first_purchase = ~df.duplicated(subset=['customer_label'], keep='first')
    df.loc[mask_first_purchase, 'purchase_diff'] = np.nan
    
    return df

df_sample_final_v6 = compute_purchase_diff_final_v6(df_sample_final_v5.copy())
df_sample_final_v6
```

```python
def cal_risk_leaving(row):
    purchase_diff = row['purchase_diff']

    if pd.isnull(purchase_diff) or (purchase_diff == 0):
        return 0
    elif 1 <= purchase_diff <= 118:
        return 1
    elif 119 <= purchase_diff <= 238:
        return 2
    elif 239 <= purchase_diff <= 365:
        return 3
    else:
        return 4
    
from tqdm import tqdm

# 'churn_risk' 컬럼 추가
df_clothes_no_new_sample['churn_risk'] = None

# 대용량 데이터에 함수 적용 (진행 상황 표시)
with tqdm(total=len(df_clothes_no_new_sample)) as pbar:
    for idx, row in df_clothes_no_new_sample.iterrows():
        churn_risk = cal_risk_leaving(row)
        df_clothes_no_new_sample.at[idx, 'churn_risk'] = churn_risk
        pbar.update(1)

df_clothes_no_new_sample.head(5)                                                                 
```
